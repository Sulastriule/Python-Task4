### Jenis-jenis Plot


##### 1. Plot Garis 


```python
from matplotlib import pyplot as plt

x = [4,6,10,13,16,20,22,24,26,30]
y = [100,120,160,180,210,270,300,320,330,350]
plt.plot(x,y)
plt.show()
```

```python
# Label dan Title

x = [4,6,10,13,16,20,22,24,26,30]
y = [100,120,160,180,210,270,300,320,330,350]
plt.plot(x,y)

# Menambah Label dan Title
plt.title("Kenaikan Subcriber berdasarkan hari")
plt.xlabel("Hari")
plt.ylabel("Jumlah Subcriber")
plt.show()
```

##### >> Legend <<

```python
# Legend
hari= [4,6,10,13,16,20,22,24,26,30]

#sumbu y pertama
solehudin = [100,120,160,180,210,270,300,320,330,350]
plt.plot(hari, solehudin)

#sumbu y kedua
sientong = [70,80,110,130,160,180,200,220,280,300]
plt.plot(hari, sientong)


# Menambah Label dan Title
plt.title("Kenaikan Subcriber berdasarkan hari")
plt.xlabel("Hari")
plt.ylabel("Jumlah Subcriber")

# Legend
plt.legend(['solehudin','sientong'])
plt.show()
```

```python
hari= [4,6,10,13,16,20,22,24,26,30]

#sumbu y pertama
solehudin = [100,120,160,180,210,270,300,320,330,350]
plt.plot(hari, solehudin, color='#444444', linestyle='--', marker='D')

#sumbu y kedua
sientong = [70,80,110,130,160,180,200,220,280,300]
plt.plot(hari, sientong, color='#5A7D9A', marker='.', linewidth='3')


# Menambah Label dan Title
plt.title("Kenaikan Subcriber berdasarkan hari")
plt.xlabel("Hari")
plt.ylabel("Jumlah Subcriber")

# Legend
plt.legend(['solehudin','sientong'])
plt.show()
```

##### 2. Plot Garis dan Bar

```python
plt.style.use('fivethirtyeight')

hari= [4,6,10,13,16,20,22,24,26,30]

#sumbu y pertama
solehudin = [100,120,160,180,210,270,300,320,330,350]
plt.bar(hari, solehudin, color='#444444', label='Solehudin')

#sumbu y kedua
sientong = [70,80,110,130,160,180,200,220,280,300]
plt.plot(hari, sientong, color='#5A7D9A', label='sientong')


# Menambah Label dan Title
plt.title("Kenaikan Subcriber berdasarkan hari")
plt.xlabel("Hari")
plt.ylabel("Jumlah Subcriber")


plt.legend(['solehudin','sientong'])
plt.grid(True)
plt.tight_layout()
plt.show()
```

##### 3. Plot Bar

```python
plt.style.use('fivethirtyeight')

hari= [4,6,10,13,16,20,22,24,26,30]

#sumbu y pertama
solehudin = [100,120,160,180,210,270,300,320,330,350]
plt.bar(hari, solehudin, color='#E5AE38', label='Solehudin')

#sumbu y kedua
sientong = [70,80,110,130,160,180,200,220,280,300]
plt.bar(hari, sientong, color='#008FD5', label='sientong')


# Menambah Label dan Title
plt.title("Kenaikan Subcriber berdasarkan hari")
plt.xlabel("Hari")
plt.ylabel("Jumlah Subcriber")


plt.legend(['solehudin','sientong'])
plt.grid(True)
plt.tight_layout()
plt.show()
```

##### 4. Plot Bar (side by side)

```python
import numpy as np

plt.style.use('fivethirtyeight')

hari= [4,6,10,13,16,20,22,24,26,30]

#menghitung posisi 
x_index = np.arange(len(hari))

#lebar bar
lebar_bar = 0.35
print(len(hari))
print(x_index)
print(x_index - lebar_bar)

#sumbu y pertama
solehudin = [100,120,160,180,210,270,300,320,330,350]
plt.bar(x_index - lebar_bar, solehudin, width=lebar_bar, color='#E5AE38', label='Solehudin')

#sumbu y kedua
sientong = [70,80,110,130,160,180,200,220,280,300]
plt.bar(x_index, sientong, width = lebar_bar, color='#008FD5', label='sientong')


# Menambah Label dan Title
plt.title("Kenaikan Subcriber berdasarkan hari")
plt.xlabel("Hari")
plt.ylabel("Jumlah Subcriber")

#xticks
plt.xticks(ticks=x_index, labels=hari)

plt.legend(['solehudin','sientong'])
plt.grid(True)
plt.tight_layout()
plt.show()
```

##### 5. Plot pie Chart

##### >> Slices <<

```python
plt.style.use('fivethirtyeight')

#slices
slices = [60,40]
plt.pie(slices)

plt.title('Contoh pie Chart')
plt.tight_layout()
plt.show()
```

##### >> Labels <<

```python
plt.style.use('fivethirtyeight')

#slices
slices = [60,40]
label = ['Enam Puluh','Empat Puluh']
plt.pie(slices, labels=label)

plt.title('Contoh pie Chart')
plt.tight_layout()
plt.show()
```

##### >> wedgeprops <<

```python
plt.style.use('fivethirtyeight')

#slices
slices = [60,40]
#labels
label = ['Enam Puluh','Empat Puluh']
# wedgeprops (garis pinggir)
plt.pie(slices, labels=label, wedgeprops = {'edgecolor':'black'})

plt.title('Contoh pie Chart')
plt.tight_layout()
plt.show()
```

##### >> Colors <<

```python
plt.style.use('fivethirtyeight')

#slices
slices = [60,40]
#labels
label = ['Enam Puluh','Empat Puluh']
#colors
color = ['#E5AE37','#609B4F']
# wedgeprops (garis pinggir)
plt.pie(slices, labels=label, colors=color, wedgeprops = {'edgecolor':'black'})

plt.title('Contoh pie Chart')
plt.tight_layout()
plt.show()
```

##### >> Explode <<

```python
plt.style.use('fivethirtyeight')

#slices
slices = [6000,4000,10000,3000,7000]
#labels
label = ['Java','SQL','Javascript','HTML/CSS','Python']
#colors
color = ['#E5AE37','#609B4F']
# Explode (pecahan pie)
explode = [0,0,0,0,0.1]
plt.pie(slices, labels=label, explode=explode, wedgeprops = {'edgecolor':'black'})

plt.title('Contoh pie Chart')
plt.tight_layout()
plt.show()
```

##### >> Shadow <<

```python
plt.style.use('fivethirtyeight')

#slices
slices = [6000,4000,10000,3000,7000]
#labels
label = ['Java','SQL','Javascript','HTML/CSS','Python']
#colors
color = ['#E5AE37','#609B4F']
# Explode (pecahan pie)
explode = [0,0,0,0,0.1]
plt.pie(slices, labels=label, shadow=True, explode=explode, wedgeprops = {'edgecolor':'black'})

plt.title('Contoh pie Chart')
plt.tight_layout()
plt.show()
```

##### >> Startangel <<

```python
plt.style.use('fivethirtyeight')

#slices
slices = [6000,4000,10000,3000,7000]
#labels
label = ['Java','SQL','Javascript','HTML/CSS','Python']
#colors
color = ['#E5AE37','#609B4F']
# Explode (pecahan pie)
explode = [0,0,0,0,0.1]
plt.pie(slices, labels=label, startangle=90, shadow=True, explode=explode, wedgeprops = {'edgecolor':'black'})

plt.title('Contoh pie Chart')
plt.tight_layout()
plt.show()
```

##### >> Autopct <<

```python
plt.style.use('fivethirtyeight')

#slices
slices = [6000,4000,10000,3000,7000]
#labels
label = ['Java','SQL','Javascript','HTML/CSS','Python']
#colors
color = ['#E5AE37','#609B4F']
# Explode (pecahan pie)
explode = [0,0,0,0,0.1]
plt.pie(slices, labels=label, startangle=90, autopct='%0.1f%%,', shadow=True, explode=explode, wedgeprops = {'edgecolor':'black'})

plt.title('Contoh pie Chart')
plt.tight_layout()
plt.show()
```
